```python
iris_setosa = df.loc[df["class"] == "Iris-setosa"];
iris_virginica = df.loc[df["class"] == "Iris-virginica"];
iris_versicolor = df.loc[df["class"] == "Iris-versicolor"];
```

```python
print("Means:")
print(np.mean(iris_setosa["petallength"]))
print(np.mean(np.append(iris_setosa["petallength"],50)));
print(np.mean(iris_virginica["petallength"]))
print(np.mean(iris_versicolor["petallength"]))
print("\nStd-dev:");
print(np.std(iris_setosa["petallength"]))
print(np.std(iris_virginica["petallength"]))
print(np.std(iris_versicolor["petallength"]))
```

```python
def hist(mean,variance,size,binsize,prior):
    class1=np.random.normal(mean,variance**0.5,size)
    class1_hist,bins=np.histogram(class1,bins=binsize)
    class1_likelihood=class1_hist/np.float64(binsize)
    class1_posterior=class1_likelihood*prior
    return class1_hist,bins,class1_likelihood,class1_posterior
```

```python
count1, bin_edges1 = np.histogram(iris_setosa['petallength'], bins=50)
count2, bin_edges2 = np.histogram(iris_virginica['petallength'], bins=50)
count3, bin_edges3 = np.histogram(iris_versicolor['petallength'], bins=50)
```

```python
likelihood1 = count1/(sum(count1))
likelihood2 = count1/(sum(count2))
likelihood3 = count1/(sum(count3))
```

```python
#print(pdf1) 
#print(bin_edges)
plt.figure(figsize=(8,6))
plt.grid()
plt.plot(bin_edges1[1:],likelihood1,color="blue", label="Iris_setosa")
plt.xlabel("petal_length")
plt.ylabel("Likelihood")
plt.legend()
plt.title("Likelihood ")
plt.show()
```

```python
#plot
plt.figure(figsize=(8,6))
plt.grid()
plt.plot(bin_edges2[1:],likelihood2,color="green",label="Iris_virginica")
plt.xlabel("petal_length")
plt.ylabel("Likelihood")
plt.legend()
plt.title("Likelihoodofvirginicaclass")
plt.show()

plt.figure(figsize=(8,6))
plt.grid()
plt.plot(bin_edges3[1:],likelihood3,color="red",label="Iris_versicolor")
plt.xlabel("petal_length")
plt.ylabel("Likelihood")
plt.legend()
plt.title("Likelihoodofversicolorclass")
plt.show()
```

```python
# finding posterior probability
pr1 = 1/3
pr2 =1/3
pr3 =1/3
post1 = pr1*likelihood1
post2 = pr2*likelihood2
post3 = pr3*likelihood3
```

```python
plt.figure(figsize=(8,6))
plt.grid()
plt.plot(bin_edges1[1:],post1,color="blue", label="Iris_setosa")
plt.xlabel("petal_length")
plt.ylabel("posterior of stosa")
plt.legend()
plt.title("posterior")
plt.show()

plt.figure(figsize=(8,6))
plt.grid()
plt.plot(bin_edges1[1:],post2,color="green", label="Iris_virginica")
plt.xlabel("petal_length")
plt.ylabel("posterior of virginica")
plt.legend()
plt.title("posterior")
plt.show()


plt.figure(figsize=(8,6))
plt.grid()
plt.plot(bin_edges1[1:],post3,color="red", label="Iris_versicolor")
plt.xlabel("petal_length")
plt.ylabel("posterior of versicolor")
plt.legend()
plt.title("posterior")
plt.show()
```
